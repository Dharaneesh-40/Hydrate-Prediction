```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Dataset

data = pd.read_csv("Daily_Water_Intake.csv")

print("Dataset loaded successfully")
```

```python
# Dataset Shape

print("Data Shape:", data.shape)
```

```python
print(data.columns)
```

```python
# Checking first few rows and summary

print(data.head())

print(data.describe(include='all'))
```

```python
## DATA CLEANING

# Checking missing values in the dataset

print("\nMissing Values per Column:")

print(data.isnull().sum())
```

```python
# Checking distribution of Hydration Level

print("\nHydration Level Count:")

print(data['Hydration Level'].value_counts())
```

```python
## VISUALIZATION OF FEATURES

# Hydration Level Distribution

plt.figure(figsize=(6,4))

sns.countplot(x='Hydration Level', data=data)

plt.title("Hydration Level Distribution")

plt.show()
```

```python
# 2. Water Intake vs Hydration Level

plt.figure(figsize=(10,6))

sns.boxplot(
    x='Hydration Level',
    y='Daily Water Intake (liters)',
    data=data
)

plt.title("Water Intake Distribution by Hydration Level")

plt.show()
```

```python
# 3. Physical Activity Level Distribution

plt.figure(figsize=(8,5))

sns.countplot(
    x='Physical Activity Level',
    data=data
)

plt.title("Physical Activity Level Distribution")

plt.show()
```

```python
# 4. Weather vs Hydration Level

plt.figure(figsize=(10,6))

sns.countplot(
    x='Weather',
    hue='Hydration Level',
    data=data
)

plt.title("Weather vs Hydration Level")

plt.show()
```

```python
## SAVING DATASET

# Exporting the cleaned data to a new CSV file

data.to_csv("Hydration_Updated.csv", index=False)

print("Processed dataset saved as Hydration_Updated.csv")
```
